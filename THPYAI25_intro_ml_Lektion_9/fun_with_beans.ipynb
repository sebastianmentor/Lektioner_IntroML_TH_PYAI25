```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
%matplotlib widget
```

```python
def pca_transformation(data):
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(data)
    principal_df = pd.DataFrame(data = principal_components
             ,columns = ['principal component 1', 'principal component 2'])
    return principal_df
```

```python
def pca_transformation_3d(data):
    pca = PCA(n_components=3)
    principal_components = pca.fit_transform(data)
    principal_df = pd.DataFrame(data = principal_components
             ,columns = ['PC 1', 'PC 2', 'PC 3'])
    return principal_df
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df.drop("ConvexArea", axis=1, inplace=True)
```

```python
inertia = []

for k in range(1, 20):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(df)
    inertia.append(km.inertia_)
```

```python
plt.plot(inertia)
```

```python
df_pca = pca_transformation(df)
```

```python
plt.scatter(df_pca["principal component 1"], df_pca["principal component 2"])
```

```python
km_no_scale = KMeans(n_clusters=2, random_state=42)
lable_no_scale = km_no_scale.fit_transform(df)
```

```python
plt.scatter(df_pca["principal component 1"], df_pca["principal component 2"], c=km_no_scale.labels_)
```
