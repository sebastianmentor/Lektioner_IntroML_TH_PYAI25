# Interaktiv 2x2-matristransformation i planet

Den här notebooken låter dig ändra en 2x2-matris
$A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$
 och direkt se hur den transformerar planet.

## Det du kan göra
- Ändra matrisens fyra element med sliders.
- Se det ursprungliga rutnätet och det transformerade rutnätet.
- Se hur enhetsvektorerna förändras.
- Slå på eller av visning av enhetskvadraten.

## Tips
- Identitetsmatrisen ger ingen förändring.
- Prova rotation, skjuvning, skalning och spegling.
- Determinanten visas i grafen och hjälper dig tolka areaförändring och orientering.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Checkbox

plt.rcParams['figure.figsize'] = (8, 8)

def draw_transformation(a=1, b=0, c=0, d=1, show_unit_square=True, show_grid=True, grid_extent=5):
    A = np.array([[a, b], [c, d]], dtype=float)
    det = np.linalg.det(A)

    # Ursprunglig vektor
    v = np.array([1, 1])

    # Transformerad vektor
    v_trans = A @ v

    # Rita ursprunglig vektor
    ax.arrow(0, 0, v[0], v[1],
            head_width=0.2,
            length_includes_head=True,
            linestyle='dashed',
            linewidth=2)

    # Rita transformerad vektor
    ax.arrow(0, 0, v_trans[0], v_trans[1],
            head_width=0.25,
            length_includes_head=True,
            linewidth=3)

    # Labels
    ax.text(v[0] + 0.1, v[1] + 0.1, "v = (1,1)")
    ax.text(v_trans[0] + 0.1, v_trans[1] + 0.1, "A v")

    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_xlim(-8, 8)
    ax.set_ylim(-8, 8)
    ax.axhline(0, linewidth=1)
    ax.axvline(0, linewidth=1)
    ax.grid(True, alpha=0.3)

    # Rita ursprungligt rutnät och transformerat rutnät
    if show_grid:
        vals = np.arange(-grid_extent, grid_extent + 1)
        t = np.linspace(-grid_extent, grid_extent, 300)

        # Ursprungliga vertikala/horizontella linjer
        for x in vals:
            ax.plot([x]*len(t), t, alpha=0.15)
        for y in vals:
            ax.plot(t, [y]*len(t), alpha=0.15)

        # Transformerade linjer
        for x in vals:
            pts = np.vstack([np.full_like(t, x), t])
            transformed = A @ pts
            ax.plot(transformed[0], transformed[1], linewidth=1.2)
        for y in vals:
            pts = np.vstack([t, np.full_like(t, y)])
            transformed = A @ pts
            ax.plot(transformed[0], transformed[1], linewidth=1.2)

    # Enhetsvektorer
    e1 = A @ np.array([1, 0])
    e2 = A @ np.array([0, 1])
    ax.arrow(0, 0, 1, 0, head_width=0.18, length_includes_head=True, linewidth=2)
    ax.arrow(0, 0, 0, 1, head_width=0.18, length_includes_head=True, linewidth=2)
    ax.arrow(0, 0, e1[0], e1[1], head_width=0.22, length_includes_head=True, linewidth=3)
    ax.arrow(0, 0, e2[0], e2[1], head_width=0.22, length_includes_head=True, linewidth=3)

    ax.text(1.1, 0.15, 'e₁', fontsize=12)
    ax.text(0.15, 1.1, 'e₂', fontsize=12)
    ax.text(e1[0] + 0.15, e1[1] + 0.15, 'A e₁', fontsize=12)
    ax.text(e2[0] + 0.15, e2[1] + 0.15, 'A e₂', fontsize=12)

    # Enhetskvadrat och dess bild
    if show_unit_square:
        square = np.array([
            [0, 0], [1, 0], [1, 1], [0, 1], [0, 0]
        ]).T
        transformed_square = A @ square
        ax.plot(square[0], square[1], linewidth=2, linestyle='--')
        ax.plot(transformed_square[0], transformed_square[1], linewidth=3)

    title = (
        f"A = [[{a:.2f}, {b:.2f}], [{c:.2f}, {d:.2f}]]    "
        f"det(A) = {det:.2f}"
    )
    ax.set_title(title)
    plt.show()


interact(
    draw_transformation,
    a=FloatSlider(value=1, min=-3, max=3, step=0.1, description='a'),
    b=FloatSlider(value=0, min=-3, max=3, step=0.1, description='b'),
    c=FloatSlider(value=0, min=-3, max=3, step=0.1, description='c'),
    d=FloatSlider(value=1, min=-3, max=3, step=0.1, description='d'),
    show_unit_square=Checkbox(value=True, description='Visa enhetskvadrat'),
    show_grid=Checkbox(value=True, description='Visa rutnät'),
    grid_extent=(2, 8, 1)
);
```

## Exempel att prova

**Identitet**
$\begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}$


**Skalning**
$\begin{bmatrix}2 & 0 \\ 0 & 0.5\end{bmatrix}$


**Skjuvning**
$\begin{bmatrix}1 & 1 \\ 0 & 1\end{bmatrix}$


**Spegling i x-axeln**
$\begin{bmatrix}1 & 0 \\ 0 & -1\end{bmatrix}$


**Rotation 90° moturs**
$\begin{bmatrix}0 & -1 \\ 1 & 0\end{bmatrix}$



```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Checkbox, IntSlider

plt.rcParams['figure.figsize'] = (8, 8)

def draw_transformation(
    a=1, b=0, c=0, d=1,
    vx=1, vy=1,
    show_unit_square=True,
    show_grid=True,
    grid_extent=5
):
    A = np.array([[a, b], [c, d]], dtype=float)
    det = np.linalg.det(A)

    v = np.array([vx, vy], dtype=float)
    v_trans = A @ v

    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_xlim(-8, 8)
    ax.set_ylim(-8, 8)
    ax.axhline(0, linewidth=1)
    ax.axvline(0, linewidth=1)
    ax.grid(True, alpha=0.3)

    # Rita ursprungligt rutnät och transformerat rutnät
    if show_grid:
        vals = np.arange(-grid_extent, grid_extent + 1)
        t = np.linspace(-grid_extent, grid_extent, 300)

        # Ursprungliga linjer
        for x in vals:
            ax.plot([x] * len(t), t, alpha=0.15)
        for y in vals:
            ax.plot(t, [y] * len(t), alpha=0.15)

        # Transformerade linjer
        for x in vals:
            pts = np.vstack([np.full_like(t, x), t])
            transformed = A @ pts
            ax.plot(transformed[0], transformed[1], linewidth=1.2)

        for y in vals:
            pts = np.vstack([t, np.full_like(t, y)])
            transformed = A @ pts
            ax.plot(transformed[0], transformed[1], linewidth=1.2)

    # Enhetsvektorer
    e1 = A @ np.array([1, 0])
    e2 = A @ np.array([0, 1])

    # Ursprungliga basvektorer
    ax.arrow(0, 0, 1, 0, head_width=0.18, length_includes_head=True, linewidth=2)
    ax.arrow(0, 0, 0, 1, head_width=0.18, length_includes_head=True, linewidth=2)

    # Transformerade basvektorer
    ax.arrow(0, 0, e1[0], e1[1], head_width=0.22, length_includes_head=True, linewidth=3)
    ax.arrow(0, 0, e2[0], e2[1], head_width=0.22, length_includes_head=True, linewidth=3)

    ax.text(1.1, 0.15, 'e₁', fontsize=12)
    ax.text(0.15, 1.1, 'e₂', fontsize=12)
    ax.text(e1[0] + 0.15, e1[1] + 0.15, 'A e₁', fontsize=12)
    ax.text(e2[0] + 0.15, e2[1] + 0.15, 'A e₂', fontsize=12)

    # Interaktiv vektor v
    ax.arrow(
        0, 0, v[0], v[1],
        head_width=0.18,
        length_includes_head=True,
        linestyle='dashed',
        linewidth=2
    )
    ax.plot(v[0], v[1], marker='o', markersize=8)
    ax.text(v[0] + 0.15, v[1] + 0.15, f'v = ({vx:.1f}, {vy:.1f})', fontsize=12)

    # Transformerad vektor Av
    ax.arrow(
        0, 0, v_trans[0], v_trans[1],
        head_width=0.22,
        length_includes_head=True,
        linewidth=3
    )
    ax.plot(v_trans[0], v_trans[1], marker='o', markersize=8)
    ax.text(
        v_trans[0] + 0.15,
        v_trans[1] + 0.15,
        f'Av = ({v_trans[0]:.2f}, {v_trans[1]:.2f})',
        fontsize=12
    )

    # Enhetskvadrat och dess bild
    if show_unit_square:
        square = np.array([
            [0, 0], [1, 0], [1, 1], [0, 1], [0, 0]
        ]).T
        transformed_square = A @ square
        ax.plot(square[0], square[1], linewidth=2, linestyle='--')
        ax.plot(transformed_square[0], transformed_square[1], linewidth=3)

    title = (
        f"A = [[{a:.2f}, {b:.2f}], [{c:.2f}, {d:.2f}]]"
        f"    det(A) = {det:.2f}"
    )
    ax.set_title(title)

    plt.show()


interact(
    draw_transformation,
    a=FloatSlider(value=1, min=-3, max=3, step=0.1, description='a'),
    b=FloatSlider(value=0, min=-3, max=3, step=0.1, description='b'),
    c=FloatSlider(value=0, min=-3, max=3, step=0.1, description='c'),
    d=FloatSlider(value=1, min=-3, max=3, step=0.1, description='d'),
    vx=FloatSlider(value=1, min=-5, max=5, step=0.5, description='vx'),
    vy=FloatSlider(value=1, min=-5, max=5, step=0.5, description='vy'),
    show_unit_square=Checkbox(value=True, description='Visa enhetskvadrat'),
    show_grid=Checkbox(value=True, description='Visa rutnät'),
    grid_extent=IntSlider(value=5, min=2, max=8, step=1, description='Rutnät')
);
```
