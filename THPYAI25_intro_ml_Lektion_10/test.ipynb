```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("fake_clusters_4_v2.csv")
```

```python
df.head()
```

```python
plt.scatter(df.x, df.y)
```

```python
inertia = []

for k in range(1, 10):
    km = KMeans(n_clusters=k)
    km.fit(df)
    inertia.append(km.inertia_)
```

```python
plt.plot([n for n in range(1,10)], inertia)
```

```python
for k in range(1, 10):
    km = KMeans(n_clusters=k)
    label = km.fit_predict(df)
    plt.scatter(df.x, df.y, c=label)
    for centroid in km.cluster_centers_:
        plt.scatter(centroid[0], centroid[1], marker='x')
    plt.show()
    print(km.inertia_)
```
