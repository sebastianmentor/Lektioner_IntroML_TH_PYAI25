```python
# imports and fun stuff
from dataclasses import dataclass
import matplotlib.pyplot as plt
```

```python
variable = 5
def add_variable(num:int) -> int:
    return num + variable

print(add_variable(9))
print(add_variable(10))
```

```python
values = [1,2,3,4,5]
s = 0
for v in values:
    s+= add_variable(v)

s
```

```python
@dataclass
class Point:
    x :int
    y :int

p1 = Point(1,2)
p2 = Point(3,-1)

print(p1, p2)
```

```python
age = [13, 21, 17, 88, 9, 44]
shoe_size = [38, 37, 42, 40, 32, 46]
hight = [120, 165, 150, 189, 100, 140]
plt.plot(age, shoe_size, 'ro', age, hight, 'bs')
plt.ylabel('shoe size')
plt.xlabel('age')
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("data.csv")

print(df.head())
```

```python
# total försäljning per produkt
df["total"] = df["pris"] * df["antal"]

# gruppera per kategori
print(df.groupby("kategori")["total"].sum())

# mest sålda produkter
print(df.groupby("produkt")["antal"].sum().sort_values(ascending=False))
```
