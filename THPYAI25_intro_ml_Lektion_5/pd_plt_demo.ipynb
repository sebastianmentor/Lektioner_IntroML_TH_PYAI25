```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Definiera vektorer
u = np.array([2, 3])
v = np.array([4, 1])

# Tabell med koordinater
df = pd.DataFrame({
    "vektor": ["u", "v"],
    "x": [u[0], v[0]],
    "y": [u[1], v[1]]
})

display(df)

# Rita vektorerna
plt.figure(figsize=(6, 6))
plt.axhline(0)
plt.axvline(0)

plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.grid(True)
plt.gca().set_aspect('equal')

plt.text(u[0] + 0.1, u[1], 'u')
plt.text(v[0] + 0.1, v[1], 'v')

plt.title("Två vektorer i planet")
plt.show()
```

```python
u = np.array([2, 3])

skalärer = [2, 0.5, -1]
vektorer = [k * u for k in skalärer]

df = pd.DataFrame({
    "skalär": skalärer,
    "x": [vec[0] for vec in vektorer],
    "y": [vec[1] for vec in vektorer]
})

display(df)

plt.figure(figsize=(6, 6))
plt.axhline(0)
plt.axvline(0)

# Originalvektor
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1)
plt.text(u[0] + 0.1, u[1], 'u')

# Skalärmultiplicerade versioner
for k, vec in zip(skalärer, vektorer):
    plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1)
    plt.text(vec[0] + 0.1, vec[1], f'{k}u')

plt.xlim(-4, 6)
plt.ylim(-4, 8)
plt.grid(True)
plt.gca().set_aspect('equal')
plt.title("Skalärmultiplikation av en vektor")
plt.show()
```

```python
u = np.array([2, 3])
v = np.array([4, 1])

df = pd.DataFrame({
    "komponent": ["x", "y"],
    "u": u,
    "v": v,
    "u*v": u * v
})

display(df)

skalärprodukt = np.dot(u, v)
print("Skalärprodukt u · v =", skalärprodukt)
```

```python
u = np.array([3, 1])

angles_deg = np.arange(0, 181, 15)
dot_products = []

for angle_deg in angles_deg:
    angle_rad = np.radians(angle_deg)
    v = np.array([np.cos(angle_rad), np.sin(angle_rad)])  # enhetsvektor
    dot_products.append(np.dot(u, v))

df = pd.DataFrame({
    "vinkel_grader": angles_deg,
    "skalärprodukt": dot_products
})

display(df)

plt.figure(figsize=(8, 5))
plt.plot(df["vinkel_grader"], df["skalärprodukt"], marker='o')
plt.axhline(0)
plt.grid(True)
plt.xlabel("Vinkel (grader)")
plt.ylabel("u · v")
plt.title("Hur skalärprodukten beror på vinkeln")
plt.show()
```

```python
vektorpar = [
    (np.array([1, 2]), np.array([2, 1])),
    (np.array([1, 0]), np.array([0, 1])),
    (np.array([2, 2]), np.array([-1, -1])),
    (np.array([3, 1]), np.array([6, 2]))
]

resultat = []

for u, v in vektorpar:
    resultat.append({
        "u": list(u),
        "v": list(v),
        "u·v": np.dot(u, v)
    })

df = pd.DataFrame(resultat)
display(df)
```

```python
u = np.array([4, 2])
v = np.array([3, 0])

proj_u_on_v = (np.dot(u, v) / np.dot(v, v)) * v

plt.figure(figsize=(7, 7))
plt.axhline(0)
plt.axvline(0)

plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, proj_u_on_v[0], proj_u_on_v[1], angles='xy', scale_units='xy', scale=1)

plt.text(u[0] + 0.1, u[1], 'u')
plt.text(v[0] + 0.1, v[1], 'v')
plt.text(proj_u_on_v[0] + 0.1, proj_u_on_v[1], 'proj av u på v')

plt.xlim(-1, 6)
plt.ylim(-1, 4)
plt.grid(True)
plt.gca().set_aspect('equal')
plt.title("Projektion med hjälp av skalärprodukt")
plt.show()
```

```python
u = np.array([2, 3])
skalärer = [-2, -1, 0, 0.5, 1, 2, 3]

resultat = []

for k in skalärer:
    ku = k * u
    resultat.append({
        "k": k,
        "ku": list(ku),
        "|ku|": np.linalg.norm(ku)
    })

df = pd.DataFrame(resultat)
display(df)

print("Originalets längd |u| =", np.linalg.norm(u))
```

```python
vektorer = [
    (np.array([1, 2]), np.array([2, -1])),
    (np.array([3, 4]), np.array([4, -3])),
    (np.array([1, 1]), np.array([1, -1])),
    (np.array([2, 3]), np.array([4, 1]))
]

resultat = []

for u, v in vektorer:
    dp = np.dot(u, v)
    resultat.append({
        "u": list(u),
        "v": list(v),
        "u·v": dp,
        "ortogonala": dp == 0
    })

df = pd.DataFrame(resultat)
display(df)
```

```python
def plot_vector(vec, label, xlim=(-5, 5), ylim=(-5, 5)):
    plt.figure(figsize=(6, 6))
    plt.axhline(0)
    plt.axvline(0)
    plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1)
    plt.text(vec[0] + 0.1, vec[1], label)
    plt.xlim(*xlim)
    plt.ylim(*ylim)
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.show()

def scalar_product(u, v):
    return np.dot(u, v)

u = np.array([2, 3])
v = np.array([1, 4])

plot_vector(u, "u")
plot_vector(v, "v")
print("u · v =", scalar_product(u, v))
```
