```python
df = pd.DataFrame({
    "dimension": ["x", "y", "z"],
    "vektor_u": [2, 4, 1]
})

skalär = 3
df["3 * u"] = df["vektor_u"] * skalär

df
```

```python

df = pd.DataFrame({
    "dimension": ["x", "y", "z"],
    "u": [2, 4, 1],
    "v": [3, -1, 5]
})

df["u * v"] = df["u"] * df["v"]
df
```

```python
skalärprodukt = df["u * v"].sum()
print("Skalärprodukt u · v =", skalärprodukt)
```

```python
df = pd.DataFrame({
    "namn": ["A", "B", "C", "D"],
    "x": [1, 2, -1, 3],
    "y": [4, 0, 2, 1]
})

df
```

```python
df = pd.DataFrame({
    "moment": ["Quiz", "Laboration", "Prov"],
    "elev_A": [8, 15, 20],
    "elev_B": [7, 18, 16]
})

df["produkt"] = df["elev_A"] * df["elev_B"]
df
```

```python
dot_product = df["produkt"].sum()
print("Skalärprodukt =", dot_product)
```

```python
def skalärprodukt(v1, v2):
    df = pd.DataFrame({
        "v1": v1,
        "v2": v2
    })
    df["produkt"] = df["v1"] * df["v2"]
    print(df)
    return df["produkt"].sum()

resultat = skalärprodukt([2, 4, 1], [3, -1, 5])
print("Skalärprodukt =", resultat)
```

```python
import numpy as np

u = np.array([2, 4, 1])
v = np.array([3, -1, 5])

df = pd.DataFrame({
    "dimension": ["x", "y", "z"],
    "u": u,
    "v": v,
    "u*v": u * v
})

display(df)

print("Skalärprodukt:", np.dot(u, v))
print("2u:", 2 * u)
```

```python
# Definiera två vektorer
u = [2, 4, 1]
v = [3, -1, 5]

df = pd.DataFrame({
    "dimension": ["x", "y", "z"],
    "u": u,
    "v": v
})

print("Originalvektorer:")
display(df)

# Skalärmultiplikation
df["2u"] = df["u"] * 2
df["-3v"] = df["v"] * -3

print("Efter skalärmultiplikation:")
display(df)

# Skalärprodukt steg för steg
df["u*v"] = df["u"] * df["v"]

print("Komponentvis multiplikation:")
display(df)

skalärprodukt = df["u*v"].sum()
print("Skalärprodukt u · v =", skalärprodukt)
```
